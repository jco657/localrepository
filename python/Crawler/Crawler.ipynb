```python
import re
import urllib.request as urllib2
import html.parser
```

```python
weburl = r'http://www.en8848.com.cn/kouyu/live/lifeenglish/144607.html' #起始网页
mp3Folder = r'E:\Workspace\IpythonNotebook\Crawler\data\mp3' #输出目录
```

```python
webHeader = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:23.0) Gecko/20100101 Firefox/23.0'}
```

```python
request = urllib2.Request(url = weburl,headers=webHeader)
webPage = urllib2.urlopen(request)
webData = webPage.read()
webData = webData.decode('utf-8')
```

```python
linkPrefix = r'http://www.en8848.com.cn'
```

```python
pageLinks= set(re.findall(r'/kouyu/live/lifeenglish/\d+.html',webData))
```

```python
print (pageLinks)
```

```python
contentLinks = []
for linksuffix in  pageLinks:
    contentLinks.append(linkPrefix+linksuffix)
print (contentLinks)
```

```python
sn = 1
contentLinksCon = contentLinks[:] #控制范围

allLink = len(contentLinksCon)

for contentLink in contentLinksCon:
    
    print ('解析：',sn,'of',allLink,',',contentLink)
    
    request = urllib2.Request(url = contentLink,headers=webHeader)
    webPage = urllib2.urlopen(request)
    webData = webPage.read()
    webData = webData.decode('utf-8')
    
    mp3Link = list(set(re.findall(r'http://mp3.en8848.com/tinglijiaocheng/kouyuhuati/qingjingkouyu/\d+.mp3',webData)))
    
    mp3Title = re.findall(r'<title>.*</title>',webData) 
    
    if len(mp3Link)==0 or len(mp3Title)==0:
        print ('音频链接或title为空，请检查！',mp3Link,mp3Title)
    
    else:
        mp3TitleRe = '生活情景英语'+mp3Title[0][mp3Title[0].index('口'):mp3Title[0].index('（')]
        print (mp3Link[0],mp3TitleRe)
        try:
            urllib2.urlretrieve(mp3Link[0],mp3Folder+'\\%s.mp3' %mp3TitleRe)
            print ('下载成功！')
        except Exception as e:
            print ('文件下载错误：',e,'执行下一任务！')
            
    webPage.close()
            
    sn += 1
```
